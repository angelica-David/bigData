```
import pandas as pd
tabela = pd.read_csv("/content/drive/MyDrive/Obesity.csv")
display(tabela)
```

```
display(tabela.info())
```

```
tabela = tabela.dropna()
display(tabela.info())
```

```
tabela = tabela.drop("CALC", axis=1)
display(tabela)
```

```
display(tabela["Obesity"].value_counts(normalize=True).map("{:.1%}".format))
display(tabela["Obesity"].value_counts())
```

```
display(tabela["SMOKE"].value_counts())
display(tabela["SMOKE"].value_counts(normalize=True).map("{:.1%}".format))
```

```
display(tabela["Obesity"].value_counts(normalize=True))

display(tabela["Obesity"].value_counts())
```

```
display(tabela["Gender"].value_counts(normalize=True))

display(tabela["Gender"].value_counts())
```

```
display(tabela["Gender"].value_counts())
display(tabela["Gender"].value_counts(normalize=True).map("{:.1%}".format))
```

```
display(tabela["family_history"].value_counts())
display(tabela["family_history"].value_counts(normalize=True).map("{:.1%}".format))
```

```
display(tabela["Age"].value_counts(normalize=True).map("{:.1%}".format))
display(tabela["Age"].value_counts())
```

```
display(tabela["MTRANS"].value_counts(normalize=True).map("{:.1%}".format))
display(tabela["MTRANS"].value_counts())
```

```
display(tabela.groupby("MTRANS",).mean(numeric_only=True))
display(tabela["MTRANS"].value_counts(normalize=True))
```

```
display(tabela.groupby("Obesity").mean(numeric_only=True))
display(tabela["Obesity"].value_counts(normalize=True))
```
